```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\MOHINI\Downloads\European_Bank.csv")
df.drop(columns=['Year', 'CustomerId', 'Surname'], inplace=True)
```

```python
df['Product_Density'] = df['NumOfProducts'] / (df['Tenure'] + 1)
```

```python
df['Engagement_Product'] = df['IsActiveMember'] * df['NumOfProducts']
```

```python
df['Balance_Salary_Ratio'] = df['Balance'] / (df['EstimatedSalary'] + 1)
```

```python
df['Age_Tenure'] = df['Age'] * df['Tenure']
```

```python
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)
```
